__a)Bank Loan Modelling__

```python
import pandas as pd
```

```python
dataset1=pd.read_excel(r"E:\Vivek\Data Science\Artificial Intelligence & Machine Learning\Notes\Day_21\Dataset\Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
dataset1
```

```python
dataset2=dataset1.drop(['ID','ZIP Code'],axis=1)
dataset2
```

```python
Y=dataset2['Personal Loan']
```

```python
X=dataset2[['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education','Mortgage', 'Securities Account', 'CD Account','Online', 'CreditCard']]
```

```python
import statsmodels.api as sm
```

```python
X1=sm.add_constant(X)
```

```python
Logistic=sm.Logit(Y,X1)
```

```python
result=Logistic.fit()
```

```python
result.summary()
```
