### Applying Chi_Square Test

```python
from scipy.stats import chi2_contingency
```

__Attrition V/s Department__

Ho : There is No Significant Difference in Department and Attrition

H1 : There is Significant Difference in Department and Attrition

```python
chitable1 = pd.crosstab(df.Attrition, df.Department)
chitable1
```

```python
stats,p,dof,expected = chi2_contingency(chitable1)
print('The Critical value is {} and p value is {}'.format(stats,p))
```

- As p value < 0.05 ,

So Null hypothesis is rejected and Alternative hypothesis is accepted ie, __There is Significant Difference in Department and Attrition__

__Attrition V/s BusinessTravel__

```python
stats,p,dof,expected = chi2_contingency(chitable2)
print('The Critical value is {} and p value is {}'.format(stats,p))
```
